```python
#import necessary lbs
import sklearn.datasets
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
breast_cancer = sklearn.datasets.load_breast_cancer() #load the data
X = breast_cancer.data 
print(X.shape,type(X))
y = breast_cancer.target
print(y.shape,type(y))

#can import in form of pandas as well
data = pd.DataFrame(X,columns=breast_cancer.feature_names)
#add y
data['class'] = y

data.head()
```

```python
data.describe() #describe the dataset
```

```python
#train-test split
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(X,y,test_size=0.15,stratify=y,random_state=0)
print(x_train.shape,x_test.shape) #see the ratio of splitting
```

```python
print(y_train.mean(),y_test.mean())
```

```python
#shows what diffrent values are possible for various columns(features)
plt.plot(x_train.T,'*')
plt.xticks(rotation='vertical')
plt.show()
#we need to do feature scaling
```

```python
from sklearn.preprocessing import Binarizer

#as it is MP neuron binarize the input between 0-1
binarizer = Binarizer()
x_train=binarizer.transform(x_train)
x_test=binarizer.transform(x_test)


#plot again
plt.plot(x_train.T,'*')
plt.xticks(rotation='vertical')
plt.show() 
## all values lie between 0-1 now
```

```python
from random import randint
```

```python
#take some example value
b = 3

i = randint(0,569) 

print("for row {}".format(i))
if (np.sum(x_train[i,:]) >= b) :
    print("Malignant")
else:
    print("Benign")

print("True value is {}".format(y_train[i]))
```

```python
#compute accuracy for a specific value

b = 2

y_pred_train = []
accurate_row = 0

for x,y in zip(x_train,y_train):
    pred = (np.sum(x)>=b)
    y_pred_train.append(pred)
    accurate_row += (y==pred) #0 if incorrect , 1 if correct

print(accurate_row/x_train.shape[0])
#we got 62% accuracy , not bad for a lowly neuron
#but dont get two proud its becuase of faulty data 
#it is predicting true if b is small and false if b has larger values
```

```python
from sklearn.metrics import accuracy_score
#on test data
y_pred_test = []

for x in x_test:
    pred_test = int(np.sum(x)>=b)
    y_pred_test.append(pred_test)
    
acc = accuracy_score(y_pred_test,y_test)

print(b,acc)
```
